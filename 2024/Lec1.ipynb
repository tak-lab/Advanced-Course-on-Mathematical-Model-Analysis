```julia
using DifferentialEquations

function SIR!(du, u, p, t)
    S, I, R = u
    β, γ = p
    N = S + I + R
    du[1] = -β*S*I/N  # dS/dt
    du[2] = β*S*I/N - γ*I  # dI/dt
    du[3] = γ*I  # dR/dt
end

# 初期値とパラメータを設定する
u₀ = [990.0, 10.0, 0.0]  # S, I, R
p = [0.45, 0.01]  # β, γ

tspan = (0.0, 200.0)  # シミュレーション時間
prob = ODEProblem(SIR!, u₀, tspan, p)

# シミュレーションを実行する
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="SIR Model", xlabel="Time (days)", ylabel="Population", label=["S" "I" "R"])
```

```julia
using DifferentialEquations

function SIRS!(du, u, p, t)
    S, I, R = u
    β, γ ,μ = p
    N = S + I + R
    du[1] = -β*S*I/N + μ*R  # dS/dt
    du[2] = β*S*I/N - γ*I  # dI/dt
    du[3] = γ*I - μ*R  # dR/dt
end

# 初期値とパラメータを設定する
u₀ = [990.0, 10.0, 0.0]  # S, I, R
p = [0.03, 0.01, 0.001]  # β, γ, μ

tspan = (0.0, 3000.0)  # シミュレーション時間
prob = ODEProblem(SIRS!, u₀, tspan, p)

# シミュレーションを実行する
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="SIR Model", xlabel="Time (days)", ylabel="Population", label=["S" "I" "R"])
```

```julia
function Lotka!(du, u, p, t)
    x, y = u
    a, b, c, d = p
    du[1] = x*(a-b*y) # dx/dt
    du[2] = y*(-c+d*x) # dy/dt
end

# 初期値とパラメータを設定する
u₀ = [90.0, 10.0]  # x, y
p = [0.2, 0.01, 0.5, 0.01]  # β, γ, μ

tspan = (0.0, 100.0)  # シミュレーション時間
prob = ODEProblem(Lotka!, u₀, tspan, p)

# シミュレーションを実行する
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="Lotka-Volterra Model", xlabel="Time (days)", ylabel="Population", label=["x" "y"])
```
