```julia
using DifferentialEquations

function SIR!(du, u, p, t)
    S, I, R = u
    ,  = p
    N = S + I + R
    du[1] = -*S*I/N  # dS/dt
    du[2] = *S*I/N - *I  # dI/dt
    du[3] = *I  # dR/dt
end

# 
u = [990.0, 10.0, 0.0]  # S, I, R
p = [0.45, 0.01]  # , 

tspan = (0.0, 200.0)  # 
prob = ODEProblem(SIR!, u, tspan, p)

# 
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="SIR Model", xlabel="Time (days)", ylabel="Population", label=["S" "I" "R"])
```

```julia
using DifferentialEquations

function SIRS!(du, u, p, t)
    S, I, R = u
    ,  , = p
    N = S + I + R
    du[1] = -*S*I/N + *R  # dS/dt
    du[2] = *S*I/N - *I  # dI/dt
    du[3] = *I - *R  # dR/dt
end

# 
u = [990.0, 10.0, 0.0]  # S, I, R
p = [0.03, 0.01, 0.001]  # , , 

tspan = (0.0, 3000.0)  # 
prob = ODEProblem(SIRS!, u, tspan, p)

# 
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="SIR Model", xlabel="Time (days)", ylabel="Population", label=["S" "I" "R"])
```

```julia
function Lotka!(du, u, p, t)
    x, y = u
    a, b, c, d = p
    du[1] = x*(a-b*y) # dx/dt
    du[2] = y*(-c+d*x) # dy/dt
end

# 
u = [90.0, 10.0]  # x, y
p = [0.2, 0.01, 0.5, 0.01]  # , , 

tspan = (0.0, 100.0)  # 
prob = ODEProblem(Lotka!, u, tspan, p)

# 
sol = solve(prob, Tsit5())

using Plots
plot(sol, title="Lotka-Volterra Model", xlabel="Time (days)", ylabel="Population", label=["x" "y"])
```
